```python
import pandas as pd
from time import time
from pathlib import Path
```

```python
SAVE_DIR = "data_output/csv/"
```

```python
num = 1045
df = pd.DataFrame(
    {
        "A": [i / 3 for i in range(num)],
        "B": [i / 7 for i in range(num)],
        "C": [i / 11 for i in range(num)],
        "D": [i / 13 for i in range(num)],
        "E": [i / 17 for i in range(num)],
        "F": [i / 19 for i in range(num)],
        "G": [i / 23 for i in range(num)],
        "H": [i / 29 for i in range(num)],
        "I": [i / 31 for i in range(num)],
        "J": [i / 37 for i in range(num)],
        "K": [i / 41 for i in range(num)],
        "L": [i / 43 for i in range(num)],
        "M": [i / 47 for i in range(num)],
    }
)
```

```python
save_dir = Path(SAVE_DIR)
save_dir.mkdir(parents=True, exist_ok=True)

total_time = time()
for i in range(1000):
    start_time = time()
    df.to_csv(save_dir / f"test_{i}.csv", index=False)
    end_time = time()
    print(f"{i}: Time taken: {end_time - start_time} seconds")

total_time = time() - total_time
print(f"Total time taken: {total_time} seconds")

```
