```python
import pandas as pd
from time import time
```

```python
num = 1045
df = pd.DataFrame(
    {
        "A": [i / 3 for i in range(num)],
        "B": [i / 7 for i in range(num)],
        "C": [i / 11 for i in range(num)],
        "D": [i / 13 for i in range(num)],
        "E": [i / 17 for i in range(num)],
        "F": [i / 19 for i in range(num)],
        "G": [i / 23 for i in range(num)],
        "H": [i / 29 for i in range(num)],
        "I": [i / 31 for i in range(num)],
        "J": [i / 37 for i in range(num)],
        "K": [i / 41 for i in range(num)],
        "L": [i / 43 for i in range(num)],
        "M": [i / 47 for i in range(num)],
    }
)
```

```python
for i in range(100):
    mode = "a" if i > 0 else "w"
    start = time()
    with pd.ExcelWriter("test.xlsx", engine="openpyxl", mode=mode) as writer:
        df.to_excel(writer, sheet_name=f"test_{i}", index=False)
    end = time()
    print(f"{i}, Time taken: {end - start} seconds")
```

```python
for i in range(100, 300):
    mode = "a" if i > 0 else "w"
    start = time()
    with pd.ExcelWriter("test.xlsx", engine="openpyxl", mode=mode) as writer:
        df.to_excel(writer, sheet_name=f"test_{i}", index=False)
    end = time()
    print(f"{i}, Time taken: {end - start} seconds")
```
