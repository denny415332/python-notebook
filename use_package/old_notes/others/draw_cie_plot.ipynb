```python
import pandas as pd
import colour
import matplotlib.pyplot as plt

from src.calc import spectrum_to_spectral_distribution
from src.draw_plot import draw_cie_xy_plot
```

```python
df = pd.read_excel("./output/25_blue_150.xlsx")
wavelengths = df.to_numpy()[:, 0]
intensities = df.to_numpy()[:, 1]
```

```python
sd = spectrum_to_spectral_distribution(wavelengths, intensities)

cie_XYZ = colour.sd_to_XYZ(sd)
cie_xyY = colour.XYZ_to_xyY(cie_XYZ)
cie_xy = colour.XYZ_to_xy(cie_XYZ)

print(f"cie_XYZ: {cie_XYZ}")
print(f"cie_xyY: {cie_xyY}")
print(f"cie_xy: {cie_xy}")
```

```python
#  CIE 1931 
fig, ax = draw_cie_xy_plot(cie_xy, "Spectrum CIE position")
ax.legend()
plt.show(fig)
```
