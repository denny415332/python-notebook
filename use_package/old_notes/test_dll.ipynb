```python
import ctypes
from pathlib import Path
```

```python
DLL_PATH = r"D:\Program Files\Ocean Optics\OceanDirect SDK\lib\OceanDirect.dll"
dll_path = Path(DLL_PATH)
```

```python
# 載入 DLL
dll = ctypes.CDLL(str(dll_path))  # 替換為實際的 DLL 檔案路徑
```

```python
import pefile

# 載入 DLL 檔案
pe = pefile.PE(str(dll_path))

fn_list = []
not_fn_list = []
# 列出匯出的函式名稱
if hasattr(pe, "DIRECTORY_ENTRY_EXPORT"):
    for export in pe.DIRECTORY_ENTRY_EXPORT.symbols:
        fn_name = "No Name"
        if export.name:
            fn_name = export.name.decode()

        if fn_name.startswith("?"):
            not_fn_list.append(fn_name)
        elif fn_name == "No Name":
            not_fn_list.append(fn_name)
        else:
            fn_list.append(fn_name)
else:
    print("未找到匯出函式")

for fn in fn_list:
    print(fn)

```
